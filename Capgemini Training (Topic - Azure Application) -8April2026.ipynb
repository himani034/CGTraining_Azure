```python
import os
from dotenv import load_dotenv
from openai import AzureOpenAI

# Load API Key securely from Colab Secrets
api_key = os.getenv("key")

# Azure resource endpoint and deployment
endpoint = "https://himani22scse1012643-863-resource.services.ai.azure.com"
deployment_name = "gpt-oss-120b"

# Initialize client with proper Azure endpoint and API version
client = AzureOpenAI(
    api_key=api_key,
    # base_url=f"{endpoint}openai/deployments/{deployment_name}/",
    # default_query={"api-version": "2024-02-01"}  # required for Azure
    api_version="2024-02-15-preview",
    azure_endpoint=endpoint
)

# Create a chat completion
completion = client.chat.completions.create(
    model=deployment_name,
    messages=[
        {
            "role": "user",
            "content": "What is the capital of France?",
        }
    ],
)

# Access the message content correctly
print(completion.choices[0].message.content)
```

```python
# Scenario
# You’re training healthcare analysts.
# You want them to see how to connect securely to Azure OpenAI, ask questions about AI in healthcare, and then deploy the interaction into a Gradio 
# interface so they can try it themselves.
```

```python
import os
from dotenv import load_dotenv
from openai import AzureOpenAI
import gradio as gr

# Load API key securely
load_dotenv()
api_key = os.getenv("key")

# Azure details (REPLACE with your correct values)
endpoint = "https://himani22scse1012643-863-resource.services.ai.azure.com"
deployment_name = "gpt-oss-120b"

# Create Azure client
client = AzureOpenAI(
    api_key=api_key,
    api_version="2024-02-15-preview",
    azure_endpoint=endpoint
)

# Function to interact with model
def healthcare_assistant(user_input):
    try:
        response = client.chat.completions.create(
            model=deployment_name,
            messages=[
                {
                    "role": "system",
                    "content": "You are a helpful healthcare AI assistant. Explain in simple terms."
                },
                {
                    "role": "user",
                    "content": user_input
                }
            ]
        )
        return response.choices[0].message.content
    
    except Exception as e:
        return f"Error: {str(e)}"

# Gradio UI
interface = gr.Interface(
    fn=healthcare_assistant,
    inputs=gr.Textbox(lines=2, placeholder="Ask healthcare question..."),
    outputs="text",
    title="Healthcare AI Assistant",
    description="Ask questions about AI in healthcare (diagnosis, treatment, etc.)"
)

# Launch app
interface.launch()
```
