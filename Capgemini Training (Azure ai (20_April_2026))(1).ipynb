```python
# Scenario
# A student is building a simple AI assistant for a class project. The assistant is
# hosted on Azure Foundry, and the student wants to ask it questions like:

# “What can you help with?”
# “Explain this topic simply.”
# “Give me a short summary.”
# Instead of talking directly to the model through a chat app, the student uses Python code to connect to the Azure service.
```

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()

api_key = os.getenv("key")
print("Loaded key:", api_key[:6] + "..." if api_key else "Missing")
my_endpoint = "https://himani123.services.ai.azure.com/api/projects/proj-default"
my_agent = "email-assistant"
my_version = "2"

api_key = os.getenv("key")

openai_client = OpenAI(
    base_url=f"{my_endpoint}/openai/v1",
    api_key=api_key
)

response = openai_client.responses.create(
    input=[{"role": "user", "content": "Tell me what you can help with."}],
    extra_body={
        "agent_reference": {
            "name": my_agent,
            "version": my_version,
            "type": "agent_reference"
        }
    },
)

print(response.output_text)
```

```python
# STUDENT PROJECT MODULE
# Build an AI Customer Support Chatbot using Azure AI Foundry
# PROJECT OBJECTIVE
# Students will:
# Create an AI chatbot for customer support
# Deploy it using Azure AI Foundry
# Integrate it with Python API
# Solve real-world customer queries
```

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()

api_key = os.getenv("key")

print("Loaded key:", api_key[:6] + "..." if api_key else "Missing")

endpoint = "https://himani123.services.ai.azure.com/api/projects/proj-default"

agent_name = "customer-support-agent"
agent_version = "2"

client = OpenAI(
    base_url=f"{endpoint}/openai/v1",
    api_key=api_key
)

messages = []

def chat(user_input):
    messages.append({"role": "user", "content": user_input})

    response = client.responses.create(
        input=messages,
        extra_body={
            "agent_reference": {
                "name": agent_name,
                "version": agent_version,
                "type": "agent_reference"
            }
        }
    )

    bot_reply = response.output_text

    messages.append({"role": "assistant", "content": bot_reply})

    return bot_reply


print("\n AI customer support Started (type 'exit' to stop)\n")

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Interview ended.")
        break

    reply = chat(user_input)
    print("Bot:", reply)
```
